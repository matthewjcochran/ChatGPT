```python
from openai import OpenAI
```

```python
client = OpenAI(
    # Defaults to os.environ.get('OPENAI_API_KEY')
    # Otherwise use: api_key='Your_API_Key'
)
```

```python
system_message = '''
Create a playlist of eight songs based on a user request.

Respond using the following format:
Playlist:
(1) <song 1>
(2) <song 2>
and so on
'''

user_message = '''
I would like to hear songs with a nonstandard time signature.
That is, I would like to hear songs that are not in 4/4.
All songs are pop, rock, or metal songs from the 60s, 70s, or 80s.
Include the time signature in parenthesis.
'''

messages = [
    {'role': 'system', 'content': system_message},    
    {'role': 'user', 'content': user_message}]

chat_completions = client.chat.completions.create(
    messages=messages,
    model='gpt-3.5-turbo',
    n=3)

playlist0 = chat_completions.choices[0].message.content
playlist1 = chat_completions.choices[1].message.content
playlist2 = chat_completions.choices[2].message.content

print(playlist0 + '\n\n' + playlist1 + '\n\n' + playlist2)
```

```python
user_message = '''
I would like to hear songs about Hawaii.
'''

messages = [
    {'role': 'system', 'content': system_message},    
    {'role': 'user', 'content': user_message}]

chat_completions = client.chat.completions.create(
    messages=messages,
    model='gpt-3.5-turbo',
    n=3)

hawaii0 = chat_completions.choices[0].message.content
hawaii1 = chat_completions.choices[1].message.content
hawaii2 = chat_completions.choices[2].message.content

print(hawaii0 + '\n\n' + hawaii1 + '\n\n' + hawaii2)
```

```python
system_message = '''
Return the exact same playlist with the year that the song was recorded in parenthesis at the end.
'''

user_message = hawaii0

messages = [
    {'role': 'system', 'content': system_message},    
    {'role': 'user', 'content': user_message}]

chat_completions = client.chat.completions.create(
    messages=messages,
    model='gpt-3.5-turbo',
    n=3)

year0 = chat_completions.choices[0].message.content
year1 = chat_completions.choices[1].message.content
year2 = chat_completions.choices[2].message.content

print(year0 + '\n\n' + year1 + '\n\n' + year2)
```

```python
system_message = '''
Each playlist contains the track number, song title, artists name, and year recorded.
Make a list of tracks where there are discrepancies  in the year recorded.

Respond using the following format:
Discrepancies:
(1) <song title 1> by <artists name 1> - <discrepancy 1 description>
(2) <song title 2> by <artists name 2> - <discrepancy 2 description>
and so on
'''

user_message = year0 + '\n\n' + year1 + '\n\n' + year2

messages = [
    {'role': 'system', 'content': system_message},    
    {'role': 'user', 'content': user_message}]

chat_completions = client.chat.completions.create(
    messages=messages,
    model='gpt-3.5-turbo',
    n=3)

discrepancy0 = chat_completions.choices[0].message.content
discrepancy1 = chat_completions.choices[1].message.content
discrepancy2 = chat_completions.choices[2].message.content

print(discrepancy0 + '\n\n' + discrepancy1 + '\n\n' + discrepancy2)
```

```python
system_message = '''
Create a simple HTML page for this playlist that inclues links to YouTube.
Put the songs in a table.
'''

user_message = year0 

messages = [
    {'role': 'system', 'content': system_message},    
    {'role': 'user', 'content': user_message}]

chat_completions = client.chat.completions.create(
    messages=messages,
    model='gpt-3.5-turbo',
    n=3,
    temperature=1)

html0 = chat_completions.choices[0].message.content
html1 = chat_completions.choices[1].message.content
html2 = chat_completions.choices[2].message.content

from IPython.display import display, HTML
display(HTML(html0 + '\n\n' + html1 + '\n\n' + html2))
```
