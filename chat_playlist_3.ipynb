```python
from openai import OpenAI
```

```python
client = OpenAI(
    # Defaults to os.environ.get('OPENAI_API_KEY')
    # Otherwise use: api_key='Your_API_Key'
)
```

```python
system_message = '''
Create a playlist of eight songs based on a user request.

Respond using the following format:
Playlist:
(1) <song 1>
(2) <song 2>
and so on
'''

user_message = '''
I'd like to hear songs with a nonstandard time signature.
That is, I'd like to hear songs that are not in 4/4.
All songs are pop, rock, or metal songs from the 60s, 70s, or 80s.
Include the time signature in parenthesis.
'''

chat_completions = client.chat.completions.create(
    messages=[
        {'role': 'system', 'content': system_message},    
        {'role': 'user', 'content': user_message}],
    model='gpt-3.5-turbo',
    n=3)

playlist0 = chat_completions.choices[0].message.content
playlist1 = chat_completions.choices[1].message.content
playlist2 = chat_completions.choices[2].message.content

print(playlist0 + '\n\n' + playlist1 + '\n\n' + playlist2)
```

```python
system_message = '''
Return the exact same playlist with the year that the song was recorded in parenthesis at the end.
'''

user_message = hawaii0

chat_completions = client.chat.completions.create(
    messages=[
        {'role': 'system', 'content': system_message},    
        {'role': 'user', 'content': user_message}],
    model='gpt-3.5-turbo',
    n=3)

year0 = chat_completions.choices[0].message.content
year1 = chat_completions.choices[1].message.content
year2 = chat_completions.choices[2].message.content

print(year0 + '\n\n' + year1 + '\n\n' + year2)
```

```python
system_message = '''
Each playlist contains the track number, song title, artists name, and year recorded.
The playlist are identical.  However, the year recorded may differ.
Make a list of tracks where there are differences in the year recorded.

Respond using the following format:
Differences:
(X) <song title X> by <artists name X> - <difference description X>
(Y) <song title Y> by <artists name Y> - <difference description Y>
and so on
'''

user_message = year0 + '\n\n' + year1 + '\n\n' + year2

messages = [
    {'role': 'system', 'content': system_message},    
    {'role': 'user', 'content': user_message}]

chat_completions = client.chat.completions.create(
    messages=messages,
    model='gpt-3.5-turbo',
    n=3)

difference0 = chat_completions.choices[0].message.content
difference1 = chat_completions.choices[1].message.content
difference2 = chat_completions.choices[2].message.content

print(difference0 + '\n\n' + difference1 + '\n\n' + difference2)
```
